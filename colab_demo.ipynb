```python
from google.colab import files
from pathlib import Path

uploaded = files.upload()
uploaded_paths = []
for name, content in uploaded.items():
    target = Path('example data') / name
    target.parent.mkdir(parents=True, exist_ok=True)
    target.write_bytes(content)
    uploaded_paths.append(str(target))

image_path = uploaded_paths[0] if uploaded_paths else 'example data/images.jpg'
print('Using image:', image_path)
```

```python
from prithvi import run_pipeline

result = run_pipeline(image_path=image_path)
result
```
